# Tune Hyperparametres

```python
from hyperopt import hp
import numpy as np
import hyperopt

def optimize_ppo(params):

    ppo = PPO(**params)

    ppo.train(1_300_000)
    score = ppo.test(100)

    return -score

space = {
    'net_width': hp.choice('net_width', np.arange(32, 256, 16, dtype=int)),
    'gamma': hp.uniform('gamma', 0.7, 0.99),
    'gae_lambda': hp.uniform('gae_lambda', 0.7, 0.99),
    'learning_rate': hp.loguniform('learning_rate', np.log(1e-5), np.log(1e-1)),
    'clip_rate': hp.uniform('clip_rate', 0.1, 0.5),
    'num_epochs': hp.choice('num_epochs', np.arange(5, 30, 3, dtype=int)),
    'steps_rollout': hp.choice('steps_rollout', np.arange(1408, 3008, 64, dtype=int)),
}

best = hyperopt.fmin(optimize_ppo, space, algo=hyperopt.tpe.suggest, max_evals=100)

print(best)
```

# Train PPO

```python

base_params = {'clip_rate': 0.2,
               'gae_lambda': 0.95,
               'gamma': 0.99,
               'learning_rate': 3e-4,
               'net_width': 64,
               'num_epochs': 10,
               'steps_rollout': 2048}

optimised_params = {'clip_rate': 0.10834440427676145,
                    'gae_lambda': 0.9039060320156153,
                    'gamma': 0.9886743327375712,
                    'learning_rate': 0.0004495484511488982,
                    'net_width': 208,
                    'num_epochs': 10,
                    'steps_rollout': 2944}

ppo = PPO(**optimised_params, save_interval=100_000_000)

ppo.train(5_000_000)
```
